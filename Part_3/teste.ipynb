```python
N=1000
x = np.linspace(-6.0, 6.0, N)
y = np.linspace(-7.0, 7.0, N)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z1 = np.exp(X * Y)
z = 50 * Z1
z[:5, :5] = -1
z = ma.masked_where(z <= 0, z)
```

```python
cs = plt.contourf(X, Y, z,
                  locator = ticker.LogLocator(),
                  cmap='bone')

cbar = plt.colorbar(cs)

plt.title('matplotlib.pyplot.contourf() Example')
plt.show()
```
